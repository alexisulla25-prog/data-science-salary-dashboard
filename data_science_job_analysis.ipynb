```python
#rename the unnamed column as 'id;
df = pd.read_csv("ds_salaries.csv")
df.head()
new_names=['id', 'work_year', 'experience_level', 'employment_type',
       'job_title', 'salary', 'salary_currency', 'salary_in_usd',
       'employee_residence', 'remote_ratio', 'company_location',
       'company_size']
df.columns= new_names
df.head()

```

```python
#check if columns types are ok!
df.dtypes
```

```python
# Salary by role

df.groupby("job_title")["salary_in_usd"].mean().round().sort_values(ascending=False)
```

```python
#Salary By experience
df.groupby("experience_level")["salary_in_usd"].mean().round().sort_values(ascending=False)
```

```python
#Salary by country
df.groupby("company_location")["salary_in_usd"].mean().round().sort_values(ascending=False)
```
